```python
import pandas as pd

df = pd.read_csv('bitcoin_2017_to_2023.csv')
```

```python
df = df.sort_values(by='timestamp', ascending=True)
```

```python
df['date'] = pd.to_datetime(df['timestamp']).dt.date
df_2 = df.drop('timestamp',axis = 1)
df_2 = df[['date', 'open', 'high', 'low', 'close', 'volume',
       'quote_asset_volume', 'number_of_trades', 'taker_buy_base_asset_volume',
       'taker_buy_quote_asset_volume']]
```

```python
# filtered_df = df_2[(df_2['date'] == '2017-08-17') ]   
# in above code I can use '2017-08-17 directly but I can't use it in below code   ?????

filtered_df = df_2.loc[df_2['date']== pd.to_datetime('2017-08-17').date()] 
# Display the result
filtered_df
```

```python
filtered_df.columns
```

```python
# Group by 'date' and apply the custom aggregations
bitcoin = df_2.groupby('date').agg({
    'open': 'first',    # Pick the first 'open' value for each group
    'high': 'max',      # Pick the maximum 'high' value for each group
    'low': 'min',       # Pick the minimum 'low' value for each group
    'close': 'last',    # Pick the last 'close' value for each group
    'volume': 'sum',     # Sum the 'volume' for each group
    'quote_asset_volume': 'sum',
    'number_of_trades': 'sum',
    'taker_buy_base_asset_volume': 'sum',
    'taker_buy_quote_asset_volume': 'sum',
    
}).reset_index()

# Display the result
bitcoin
```

```python
# Create a new column 'price_increase' to store 1 if the close price is higher than the previous day, 0 otherwise
bitcoin['price_increase'] = (bitcoin['close'] < bitcoin['close'].shift(-1)).astype(int)

# Display the DataFrame with the new column
bitcoin
```

```python
bitcoin.to_csv('My_dataframe.csv', index=False)  # Save to CSV without the index
```
